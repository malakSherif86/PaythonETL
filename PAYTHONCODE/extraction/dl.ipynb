```python
import pandas as pd
from datetime import datetime

def add_columns_and_export(source_path, output_path):
  
    df = pd.read_csv(source_path, header=0)
    
    df['extract_time'] = datetime.now()
    
   
    df['source_name'] = source_path
  
    df.to_csv(output_path, index=False)


source_path = '/home/malak/bikestore_dim/staff/staffs.csv'
output_path = '/home/malak/landing/staff/staff.csv'
add_columns_and_export(source_path, output_path)
```

```python
add_columns_and_export('/home/malak/bikestore_dim/customers/customers.csv', '/home/malak/landing/customers/customers.csv')
```
