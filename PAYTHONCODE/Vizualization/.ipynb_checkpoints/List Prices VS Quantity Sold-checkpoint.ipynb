```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/home/malak/MODLING/information_mart/orders_products_details.csv')
```

```python


scatter_color = '#aca6b3'

sns.jointplot(x='list_price_x', y='quantity', data=df, kind='scatter', color=scatter_color, height=8)
plt.title('                          List Prices vs Quantity Sold')
plt.xlabel('List Price')
plt.ylabel('Quantity Sold')
plt.show()
```
