```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/home/malak/MODLING/information_mart/orders_products_details.csv')
```

```python


brand_distribution = df['brand_id'].value_counts().reset_index()
brand_distribution.columns = ['Brand ID', 'Number of Orders']

plt.figure(figsize=(10, 6))
sns.barplot(x='Brand ID', y='Number of Orders', data=brand_distribution)
plt.title('Distribution of Orders Across Different Brands')
plt.show()
```
