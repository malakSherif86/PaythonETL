```python
import pandas as pd

df = pd.read_csv('/home/malak/landing/orders/orders.csv', header=0)
```

```python
df.info()
```

```python
import pandas as pd

date_columns = ['order_date', 'shipped_date', 'required_date']
for column in date_columns:
    df[column] = pd.to_datetime(df[column], errors='coerce') 
    df[column] = df[column].apply(lambda x: x if 2000 <= x.year <= 2024 else pd.NaT) 

print(df.head())
```

```python
df
```

```python

null_values = df.isnull()

null_counts = null_values.sum()

any_null = null_values.any()

print("Null counts per column:")
print(null_counts)
print("\nAny null values per column:")
print(any_null)
```

```python
duplicate_rows = df.duplicated()
print(df[duplicate_rows])
```

```python
df = df.drop_duplicates()
```

```python
duplicates_exist = df.duplicated().any()
```

```python
if not duplicates_exist:
    print("No duplicate rows exist in the cleaned DataFrame.")
else:
    print("Duplicate rows still exist in the cleaned DataFrame.")
```

```python
df
```

```python
# Fill null values with forward fill (value from the previous row)
df = df.fillna(method='ffill')
```

```python
df['extract_time'] = pd.to_datetime(df['extract_time'])
```

```python
df.to_csv('/home/malak/staging_1/orders/orders.csv', index=False)
```
