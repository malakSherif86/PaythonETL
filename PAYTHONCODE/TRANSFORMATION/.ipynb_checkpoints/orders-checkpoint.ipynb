```python
import pandas as pd

df = pd.read_csv('/home/malak/staging_1/orders/orders.csv', header=0)
```

```python
df.info()
```

```python
import pandas as pd
df['required_date'] = pd.to_datetime(df['required_date'])
df['shipped_date'] = pd.to_datetime(df['shipped_date'])
df['late_delivery'] = df['shipped_date'] > df['required_date']
df.info()
```

```python
df
```

```python
df['latency_days'] = df['shipped_date'] - df['required_date']
df['latency_days'] = df['latency_days'].apply(lambda x: x.days if x.days > 0 else pd.NaT)
```

```python
df
```

```python
df['weekend'] = df['shipped_date'].dt.dayofweek.isin([0, 6])
```

```python
import pandas as pd

order_status_lookup = pd.DataFrame({
    'order_status': [1, 2, 3, 4],
    'status_description': ['Pending', 'Processing', 'Rejected', 'Completed']
})

df = pd.merge(df, order_status_lookup, how='left', on='order_status')
df.drop(columns=['order_status'], inplace=True)
df.rename(columns={'status_description': 'orderStatus'}, inplace=True)
print(df)
```
