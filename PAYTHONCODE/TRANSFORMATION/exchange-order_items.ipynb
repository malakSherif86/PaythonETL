```python
import pandas as pd

order_items_df = pd.read_csv('/home/malak/staging_1/order_items/order_items.csv')
exchange_df = pd.read_csv('/home/malak/staging_1/exchange/exchange.csv', header=None, nrows=2)
value = exchange_df.iloc[1, 0]
usd_to_egp_df = pd.DataFrame({'USD_to_EGP': [value]})
OI = pd.merge(order_items_df.assign(key=1), usd_to_egp_df.assign(key=1), on='key').drop('key', axis=1)
print(OI.head())
#used drop key to drop the key for the exchange , without that there would be a key column that will be repated all over the table 
```

```python

OI['USD_to_EGP'] = OI['USD_to_EGP'].astype(float)

OI['list_price_egp'] = OI['list_price'] * OI['USD_to_EGP']
```
